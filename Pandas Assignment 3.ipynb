# 30.Create a new column rating_category using apply()

```python
def rating_category(x):
    if x >= 8:
        return "Excellent"
    elif x >= 6:
        return "Good"
    else:
        return "Average"
df['rating_category'] = df['imdb_rating'].apply(rating_category) 
```
